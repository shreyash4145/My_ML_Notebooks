## Step 3: Data Preparation

- As we observed, column dteday is DATE col but reflecting as an object.

`Action Requested --> Typecasting`

- In order to fit a regression line, we would need numerical values for columns seasons, mnth and weathersit to 1s and 0s, where 1 is a 'Yes' and 0 is a 'No'.

`Action Requested --> 1) Convert categorical columns to have actual values and 2) Perform OHE of the categorical cols`

- Index col 'instant' and 'dteday' are not required

`Action Requested --> Drop them`

```python
# Changing data type of dteday

bk_shrng_df.dteday = bk_shrng_df.dteday.astype('datetime64')
bk_shrng_df.info()
```

### Fixing season column

```python
bk_shrng_df.season.value_counts()
```

```python
# Step 1: Convert categorical columns to have actual values

bk_shrng_df.season = bk_shrng_df.season.apply(lambda x: 'spring' if x==1 else 'summer' if x==2 else 'fall' if x==3 else 'winter')
bk_shrng_df.head()
```

```python
bk_shrng_df.season.value_counts()
```

```python
# Step 2: Perform OHE
#The variable `season` has four levels. We need to convert these levels in the form of 0s and 1s. 
#For this, we will use something called `dummy variables`.

season_cnvrt = pd.get_dummies(bk_shrng_df.season)
season_cnvrt.head()
```

Now, we don't need four columns. we can drop the `fall` column, as the type of season can be identified with just the last three columns where — 
- `000` will correspond to `fall`
- `001` will correspond to `spring`
- `010` will correspond to `summer`
- `011` will correspond to `winter`

```python
# Now let's drop the first column from season_cnvrt df using 'drop_first = True'

season_cnvrt = pd.get_dummies(bk_shrng_df.season, drop_first = True)
season_cnvrt.head()
```

```python
# We will add the results to Oiginal datafram

bk_shrng_df = pd.concat([bk_shrng_df, season_cnvrt], axis = 1)
bk_shrng_df.head()
```

```python
# Drop 'season' as we have created the dummies for it

bk_shrng_df.drop(['season'], axis = 1, inplace = True)
bk_shrng_df.head()
```

### Fixing 'mnth' column

```python
bk_shrng_df.mnth.value_counts()
```

### Rescaling the Features 

As we learnt in the demonstration of Simple Linear Regression, scaling doesn't impact your model. Here we can see that except for `area`, all the columns have small integer values. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling 
2. Standardisation (mean-0, sigma-1) 

This time, we will use MinMax scaling.

### Checking VIF

Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model. The formula for calculating `VIF` is:

### $ VIF_i = \frac{1}{1 - {R_i}^2} $

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

- The common heuristic we follow for the VIF values is:

`> 10:  VIF value is definitely high, and the variable should be eliminated.`

`> 5:  Can be okay, but it is worth inspecting.`

`< 5: Good VIF value. No need to eliminate this variable.`

### Dropping the variable  based on p-values and updating the model

As we can see from the summary and the VIF dataframe, some variables are still insignificant. We will follow below table for dropping columns.

    p-value      VIF              Action_Requested
    
    High         High             `REMOVE it`
    High         Low              `First remove the cols having High p-val and Low VIF`
    Low          High             `Then remove cols having High VIF and Low p-val`
    Low          Low              `KEEP the column`

`Its always advisable and best practice to DROP only one column at a time and observe the model behavior.`

`We will start with the highest p-val (insignificant var) 0.787 and then drop 1 by 1`
