```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
#Reading Data
data=pd.read_csv('headbrain.csv')
print(data.shape)
data.head()

```

```python
#We are finding linear relationship betwwen headsize and BrainWeight
#Collecting X and Y
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
#To find the value of m &c as Y=mX+c
#Mean X and Y
mean_X=np.mean(X)
mean_Y=np.mean(Y)

#Total No.of Values
n=len(X)
print(n)

#Using formula to calculate m & c
numer=0
denom=0
for i in range(n):
    numer += (X[i] - mean_X)*(Y[i] - mean_Y)
    denom += (X[i] - mean_X) ** 2
m=numer/denom
c=mean_Y - (m*mean_X)

#Print Coefficients
print(m,c)
                                                                                        
```

```python
#Ploting Values and Regression Line

max_X=np.max(X)+100
min_X=np.min(X)-100

# Calculating line values X and Y

X=np.linspace(min_X,max_X,1000)
Y=c+m*X

# Plotting Line
plt.plot(X,Y,color='#58b970' , label='Regressin Line')
#ploting Scatter Point
plt.scatter(X,Y,c='#ef5423', label='Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in gram')
plt.legend()
plt.show()

```

```python
ss_t=0
ss_r=0
for i in range(n):
    y_pred=c+m*X[i]
    ss_t+=(Y[i]-mean_Y) **2
    ss_r+=(Y[i]-y_pred) ** 2
r2=1-(ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X=X.reshape((len(X),1))

#Creating Model
reg=LinearRegression()

#Fitting training data
reg=reg.fit(X,Y)

# Y Prediction
Y_pred=reg.predict(X)
#print(Y_pred)

#Calculating RMSE and R2 Score
mse=mean_squared_error(Y,Y_pred)
rmse=np.sqrt(mse)
r2_score=reg.score(X,Y)
print(np.sqrt(mse))
print(r2_score)
```
