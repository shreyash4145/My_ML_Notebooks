## Underfitting model
### We have trained the Linear Regression model using the sales dataset. Below we have seperated out some data for training and testing and shown that the trained model is having High Bias and Low Variance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('Sales_Prediction_1.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

x_testing=[X.pop(3),X.pop(5)];
y_testing=[Y.pop(3),Y.pop(5)];

print("Data:-")
print(data)
print("")
print("Testing Data:-")
print(x_testing[0],y_testing[0])
print(x_testing[1],y_testing[1])
```

```python
# Building the model
m = 0
c = 0

L = 0.01  # The learning Rate
epochs = 10  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent
for i in range(epochs):
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c

# Making predictions

Y_pred = m*X + c
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='yellow')  # regression line
plt.show()
```

### We can see above that the Bias is high and Below we have plotted the testing dataset, we can see that the variance is low for it.

```python
plt.scatter(x_testing, y_testing)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='yellow')  # regression line
plt.show()
```
