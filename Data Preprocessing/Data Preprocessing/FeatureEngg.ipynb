```python
#Categorical Encoding: Count Encoding
#Convert each categorical value to its frequency(the number of times it appears in the dataset).
dataset = pd.read_csv('penguins_size.csv')
print(dataset)


```

```python
categorical_data = dataset.drop(['culmen_length_mm', 'culmen_depth_mm','flipper_length_mm', 'body_mass_g'], axis=1)
```

```python
#detecting and removing outliers using standard deviation
```

```python


factor = 2
upper_lim = dataset['culmen_length_mm'].mean () + dataset['culmen_length_mm'].std () * factor
lower_lim = dataset['culmen_length_mm'].mean () - dataset['culmen_length_mm'].std () * factor

no_outliers = dataset[(dataset['culmen_length_mm'] < upper_lim) & (dataset['culmen_length_mm'] > lower_lim)]
no_outliers

```
